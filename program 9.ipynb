```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Generate some random data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([0, 0, 0, 1, 1, 1])

# Plot the data
plt.scatter(x, y)
plt.title('Scatter plot of data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# Create a Logistic Regression model and fit it to the data
model = LogisticRegression(solver='lbfgs').fit(x.reshape(-1, 1), y)

# Print the model coefficients and intercept
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)

# Predict the y-values for some new x-values
new_x = np.array([15, 1, 2])
new_y = model.predict(new_x.reshape(-1, 1))
print('Predicted y-values:', new_y)

# Plot the data with the logistic regression curve
plt.scatter(x, y)
plt.plot(x, model.predict_proba(x.reshape(-1, 1))[:,1], color='red')
plt.title('Logistic Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
